The Traffic and Data Related Violations Dataset consists of multiple rows(52966) of traffic stops and 15 columns. Brief overview of the key columns:

* __stop_date:__ Date of violation

* __driver_age:__ Age of violators
* __driver_race:__ Race of violators
* __violation:__ Categories of violations: Speeding, Moving Violation (Reckless Driving, Hit and run, Assaulting another driver, pedestrian, improper turns and lane changes, etc),Equipment (Window tint violations, Headlight/taillights out, Loud exhaust, Cracked windshield, etc.), Registration/Plates,Seat Belt, or other (Call for Service, Violation of City/Town Ordinance, Suspicious Person, Motorist Assist/Courtesy, etc.)

* __stop_outcome:__ Result of violation
* __is_arrested:__ Whether a person was arrested in True or False form
* __stop_duration:__ Detained time for violators approx(in minutes)
* __drugs_related_stop:__ Whether a person was involved in drug crime(True,False)

```python
# Copy the original DataFrame to a new DataFrame
df_new = df.copy()
df_new
```

```python
df = df.drop_duplicates()
df = df.drop_duplicates(keep=False)
print(df.copy)
```

df = df.drop_duplicates() removes exact duplicate rows across our dataset.

```python
rows_with_3_or_more_empty = df_new.isnull().sum(axis=1) >= 3
df_cleaned = df_new[~rows_with_3_or_more_empty]
df_new = df_cleaned.drop(columns = ['country_name'])

df_new.isnull().sum()
```

For this code I dropped the rows that had 3 or more missing values. 
df_new.isnull() creates a boolean DataFrame where True represents a missing value (NaN).
.sum(axis=1) counts the number of True values (i.e., missing values) in each row.
3 creates a boolean Series, where True means the row has 3 or more missing values.
~rows_with_3_or_more_empty inverts the boolean mask, selecting rows with fewer than 3 missing values.
I then dropped the country name column because there were no values in the column at all therefore it was not needed for visualization.
df_new.isnull().sum() shows the count of missing values in each column. All of the columns now have 0 except driver_age and search_type

```python
df_new.loc[df_new['search_type'].isna(), 'search_type'] = "Search not conducted"
```

```python
df_new.isnull().sum()
```

### Part 5: Analysis

```python
#Investigate whether certain driver demographics (e.g., gender, race) are disproportionately associated with specific violations.
contingency_table = pd.crosstab([df_new['driver_gender'], df_new['driver_race']], df_new['violation'], dropna=False)
contingency_table_normalized = contingency_table.div(contingency_table.sum(axis=1), axis=0)
print(contingency_table_normalized)
```

To investigate whether certain driver demographics (e.g., gender, race) are disproportionately associated with specific violations, I decided to create a contingency table with counts of violation for each combination of driver_gender and driver_race.
div()): This normalizes the counts by dividing by the sum of counts per row (demographic group), so the result will be proportions rather than raw counts. 
Based on this contingency table, you can tell that the proportions are higher all under the speeding column. You could say that there are many disproportionately associations between the races and the violations because they are way too small. 

```python
# Plot a heatmap of the normalized values
plt.figure(figsize=(12, 8))
sns.heatmap(contingency_table_normalized, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)

plt.title('Proportions of Violations by Gender and Race')
plt.xlabel('Violation Type')
plt.ylabel('Driver Demographics (Gender and Race)')
plt.tight_layout()
plt.show()
```

I also decided to plot a heatmap as well to show whether certain driver demographics (e.g., gender, race) are disproportionately associated with specific violations. The heatmap also shows what I was saying in the contingency table about speeding having the highest correlations with driver gender and driver race. 

```python
# Create a contingency table to show counts of each combination of 'search_conducted' and 'is_arrested'
contingency_table = pd.crosstab(df_new['search_conducted'], df_new['is_arrested'], dropna=False)

# Display the contingency table
print(contingency_table)
```

Earlier during the data visualization section we already examined these two categories in a stacked bar plot so I decided to use a contingency table this time. It shows us that most of the traffic stops didnt conduct a search and wasn't arrested. On the other hand, a search being conducted along with not being arrested is not common in the dataset.

```python
# Count the occurrences of each gender
import matplotlib.patches as mpatches
gender_counts = df_new['driver_gender'].value_counts()
gender_counts = gender_counts[['Male', 'Female']] if 'Male' in gender_counts and 'Female' in gender_counts else gender_counts.sort_index(ascending=False)
# Create a pie chart with the gender counts
fig, ax = plt.subplots(figsize=(8, 8))
# Define colors for Male and Female
colors = ['#66b3ff', '#ff6666']  # Blue for Male, Red for Female
ax.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
ax.set_title(f"Gender of People", fontsize=16)
# Create custom legend for Male and Female colors
male_patch = mpatches.Patch(color='#66b3ff', label='Male')
female_patch = mpatches.Patch(color='#ff6666', label='Female')
# Add the legend to the plot (hue for Male and Female on the side)
ax.legend(handles=[male_patch, female_patch], loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12)
# Display the pie chart
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()  # Adjust layout to avoid clipping the legend
plt.show()
```

This visualizaiton is a pie chart showing the distribution of the gender of  people. palette=['#66b3ff', '#ff6666']: This specifies the colors for Male and Female (blue for Male, red for Female).
